## Background and Prior Work

LEGO® Group was built brick by brick to become the powerhouse that it is today, reaching millions of consumers and accumulating billions in revenue each year. LEGO®s were not always synonymous with the versatile plastic pieces meant for play, exploration, art or more. The next few decades witness an explosion in available LEGO® sets due to the introduction of new pieces, themes, and entertainment as well as construction of stores, factories and theme parks. 

A 2016 analysis conducted by Mode data scientist Joel Carron, revealed some interesting findings from a 67-year timeline of LEGO® sets. This is not surprising as the whole world was experiencing a financial crisis and many companies struggled to stay afloat. Hence, this period of financial uncertainty could have an impact on our data, and we need to be cognizant of the fact.

```python
# import libraries
import numpy as np
import pandas as pd
import re
from urllib.request import Request, urlopen
```

### Colors
We will combine colors data from two sources: Rebrickable and BrickLink
We will first handle BrickLink. Since the data is taken from a webpage that uses strange HTML, we will need to do a lot of wrangling to get the raw data into a usable form. We will not be using Modulex colors since Modulex is a separate thing,

```python
# Pull color data
req = Request(
    url='https://www.bricklink.com/catalogColors.asp?v=0&itemType=P&itemNo=', 
    headers={'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:122.0) Gecko/20100101 Firefox/122.0'}
)
color_bricklink_tables = pd.read_html(urlopen(req))

# Helper functions
def extract_first_date(string):
    string = string.replace('\xa0','')
    match = re.match(r'(\d+)-(\d+)', string)
    if match == None:
        return np.nan
    else:
        return match.group(1)

def extract_last_date(string):
    string = string.replace('\xa0','')
    match = re.match(r'(\d+)-(\d+)', string)
    if match == None:
        return np.nan
    else:
        return match.group(2)

def bricklink_fix(df):
    df = df.dropna(axis=1, how='all')
    df.columns = df.iloc[0]
    df = df.drop(df.index[0])
    #print(df.head())
    df = df.set_index('ID')
    #df['Color Timeline'].apply(extract_dates)
    df['First Year Active'] = df['Color Timeline'].apply(extract_first_date)
    df['Recent Year Active'] = df['Color Timeline'].apply(extract_last_date)
    df = df.drop("Color Timeline", axis=1)
    return df    

# Fix all of the tables and combine them
bricklink_tables = []
bricklink_tables.append(bricklink_fix(color_bricklink_tables[4]))
bricklink_tables.append(bricklink_fix(color_bricklink_tables[7]))
bricklink_tables.append(bricklink_fix(color_bricklink_tables[10]))
bricklink_tables.append(bricklink_fix(color_bricklink_tables[13]))
bricklink_tables.append(bricklink_fix(color_bricklink_tables[16]))
bricklink_tables.append(bricklink_fix(color_bricklink_tables[19]))
bricklink_tables.append(bricklink_fix(color_bricklink_tables[22]))
bricklink_tables.append(bricklink_fix(color_bricklink_tables[25]))
bricklink_tables.append(bricklink_fix(color_bricklink_tables[28]))
bricklink = pd.concat(bricklink_tables)
bricklink = bricklink.dropna()
bricklink.head()
```

We will now get the data from Rebrickable. We will drop obscure or unreleased colors from the set. Since we did Bricklink, we will use the BrickLink id for the index. We then merge the two datasets into our final color DataFrame 

```python
def get_bricklink_id(string):
    string = string.strip()
    match = re.match(r'(\d+)', string)
    return match.group(1)

colors_detailed = pd.read_csv('rebrickable/colors_updated.csv')
# These colors do not have official names or have any dates. These colors are extremely rare so we probably dont' need them.
colors_detailed = colors_detailed.dropna(axis=0, subset=['LEGO', 'First Year', 'BrickLink'])
colors_detailed = colors_detailed.drop(['Img', 'Unnamed: 0', 'ID', 'First Year', 'Last Year', 'Name'], axis=1)
colors_detailed['BrickLink ID'] = colors_detailed['BrickLink'].apply(get_bricklink_id)
colors_detailed = colors_detailed.set_index('BrickLink ID')
colors_detailed

colors_df = pd.merge(colors_detailed, bricklink, left_index=True, right_index=True)
colors_df.head()
```

### Sets, Themes, and Minifigures
Here we will combine set, minifigure, and theme data from Rebrickable.We first cleaned up the set information, dropping products that are not sets.

```python
themes = pd.read_csv('rebrickable/themes.csv')
themes = themes.set_index('id')

themes['parent_id'] = themes['parent_id'].apply(lambda num: int(num) if not pd.isna(num) else np.nan)

sets = pd.read_csv('rebrickable/sets.csv')
sets = sets.drop('img_url', axis=1)
# Most, if not all, of these are either not lego sets (eg books or video games), bundles of other sets, or single replacement parts.
sets = sets.drop(sets[sets.num_parts==0].index, axis = 0)
sets = sets.drop(sets[sets.num_parts==1].index, axis = 0)
# Replaces the Theme ID with the theme name and subtheme (if there is one)
sets['theme'] = sets['theme_id'].apply(lambda string: themes['name'].loc[string] if np.isnan(themes['parent_id'].loc[string]) else themes['name'].loc[themes['parent_id'].loc[string]])
sets['subtheme'] = sets['theme_id'].apply(lambda string: np.nan if np.isnan(themes['parent_id'].loc[string]) else themes['name'].loc[string] )
sets = sets.drop('theme_id', axis = 1)

# Drop things that are not sets
sets = sets.drop(sets[sets.theme=='Books'].index, axis=0)
sets = sets.drop(sets[sets.theme=='Gear'].index, axis=0)
sets.head()
```

We then started the process of combining the datasets. Rebrickable's dataset is based around inventories. Each inventory corresponds to a Lego product. Each inventory can contain parts, sets, or minifigures. These are separated into three csv files, linking each part/set/minifigure to its inventory ID. There is another csv that contains all inventory IDs and the set it belongs to.

Inventories that contain sets are usually collections of sets, so we can drop them. 

```python
# Read the files and drop irrelevant columns
inventories = pd.read_csv('rebrickable/inventories.csv')
inventories = inventories.drop('version', axis = 1)

inventory_minifigs = pd.read_csv('rebrickable/inventory_minifigs.csv')

minifigs = pd.read_csv('rebrickable/minifigs.csv')
minifigs = minifigs.drop('img_url', axis=1)

inventory_sets =  pd.read_csv('rebrickable/inventory_sets.csv')
inventory_sets = inventory_sets.set_index('inventory_id')

inventories = inventories.set_index('id')
# drop collections of sets.
inventories = inventories.drop(inventory_sets.index, errors='ignore')
inventories.head()
```

We then converted the the inventory of minifigures into lists of figure numbers and quantity pairs and then added this to `inventories`. We then merged the data.

```python
# This will collect all minifigures in the same inventory and put them into a list of (fig_num, quantity) tuples
minifig_dict = dict()

for index, row in inventory_minifigs.iterrows():
    if row['inventory_id'] not in minifig_dict:
        minifig_dict[row['inventory_id']] = []
    minifig_dict[row['inventory_id']].append((row['fig_num'], row['quantity']))

inventories['fig_info'] = pd.Series(minifig_dict)

# merge everything
inventory_df = pd.merge(inventories, sets, on='set_num')
inventory_df.head()
```

We will define a function to clean Launch_and_Exit column and apply it to the designated column.
 
We will also separate the Launch_and_Exit column by splitting it into two new columns based on separated dates. Then we will delete the original Launch_and_Exit column.

```python
#Define a function to clean 'Launch_and_Exit' column 

def clean_LnE(string):

    new_string = re.sub(r'\([^)]*\)', '', string)

    return new_string
```

```python
#Apply the function to designated column 

brickset['Launch_and_Exit'] = brickset['Launch_and_Exit'].astype('str')
brickset['Launch_and_Exit'] = brickset['Launch_and_Exit'].apply(clean_LnE)
```

```python
#Separate launch and exit date 

for date in brickset['Launch_and_Exit']:
  sep_string = re.split(" - ", date)
```

```python
#create two new columns based on separated dates. Then, delete original "Launch_and_Exit" column

brickset['Launch_Date'] = sep_string[0]

brickset['Exit_Date'] = sep_string[1]

brickset = brickset.drop(columns = 'Launch_and_Exit')
```
